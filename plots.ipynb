```python
from src.models.entry_model import get_entries, get_entry, create_entry
from src.models.co_model import get_co_by_entry_id, get_co_entries, create_co_entry
from src.models.pm10_model import get_pm10_by_entry_id, get_pm10_entries, create_pm10_entry
from src.models.pm25_model import get_pm25_by_entry_id, get_pm25_entries, create_pm25_entry
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
columns=['datetime', 'co_value', 'co_description', 'co_qualification', 'pm10_value', 'pm10_description', 'pm10_qualification', 'pm25_value', 'pm25_description', 'pm25_qualification']
df = pd.DataFrame(columns=columns)
entries = get_entries()
for entry in entries:
    co = get_co_by_entry_id(entry.id)
    pm10 = get_pm10_by_entry_id(entry.id)
    pm25 = get_pm25_by_entry_id(entry.id)
    df.loc[len(df.index)] = [entry.recording_date, co.value, co.description, co.qualification, pm10.value, pm10.description, pm10.qualification, pm25.value, pm25.description, pm25.qualification]
```

```python
plt.figure(figsize=(10,6))

plt.title('Pollutants quantity plot') 

plt.plot(df['datetime'], df['co_value'], label='co')
plt.plot(df['datetime'], df['pm10_value'], label='pm10')
plt.plot(df['datetime'], df['pm25_value'], label='pm2.5')
    
plt.ylabel('ug/m^3') 
  
plt.legend() 

plt.show()
```
